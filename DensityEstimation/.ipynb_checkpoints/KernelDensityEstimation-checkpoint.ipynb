# <center> <u> Kernel Density Estimation </u> </center> 

```python
import numpy as np
import pandas as pd 

from numba import jit, prange
from scipy import stats

import seaborn as sns
import matplotlib.pyplot as plt
```

## <center> Set Seaborn Plot Theme </center> 

```python
sns.set_theme(
    context = "notebook",
    style = "darkgrid", 
    palette = "bright",
)

sns.set(rc = {'figure.figsize':(10, 5)})
```

## <center> Generate Synthetic Data </center>

```python
N = 10000
rv = np.random.gumbel(0, 1, N)
```

```python
ax = sns.histplot(rv)

ax.set(
    title = "Gumble Distribution",
)

plt.show()
```

## <center> Kernel Density Estimators </center> 

## <center> $ \hat{f_{h}(x)} = \frac{1}{nh} \sum_{i=1}^{n} K(\frac{x - x_{i}}{h}) $ </center>

## <center> $ K(u) = \frac{1}{\sqrt{2 \pi}} e^{-\frac{1}{2} u^2} $ </center>

```python
@jit(nopython = True)
def gaussian_kernel(u):
    """
    
    """
    
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * u**2)
```

## <center> $ K(u) = \frac{1}{2} e^{- |x| } $ </center>

```python
@jit(nopython = True)
def laplace_kernel(u):
    """
    
    """
    
    return (1 / 2) * np.exp(-np.abs(u))
```

```python
@jit(nopython = True, parallel = True)
def compute_kernel_estimators(h, arr):
    """
    
    """
        
    N = len(arr)
    gaussian_kde = np.zeros(N)
    laplacian_kde = np.zeros(N)
    
    for i in prange(N):
        
        gaussian_total = 0
        laplacian_total = 0
        x = arr[i]
        
        for j in prange(N):
            
            u = (x - arr[j]) / h
            gaussian_total += gaussian_kernel(u)
            laplacian_total += laplace_kernel(u)
        
        gaussian_kde[i] = (1 / (N * h)) * gaussian_total
        laplacian_kde[i] = (1 / (N * h)) * laplacian_total
    
    return gaussian_kde, laplacian_kde
```

```python
h = 1.06 * np.std(rv) * len(rv)**(-1/5)
print("Smoothing Parameter: ", h)

sorted_x = np.sort(rv)
gaussian_kde, laplacian_kde = compute_kernel_estimators(h, sorted_x)
```

```python
ax = sns.lineplot(
    x = sorted_x,
    y = gaussian_kde,
    label = "Theoretical Gaussian Density",
)

sns.lineplot(
    x = sorted_x,
    y = laplacian_kde,
    label = "Theoretical Laplacian Density",
)

ax.set(
    title = "Kernel Density Estimation",
)

plt.show()
```
