# <center> <u> Loss Functions for Classification </u> </center>  

```python
import numpy as np

import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
P = np.linspace(1e-7, 1, 1000)
```

## <center> Entropy Impurity </center>

## <center> $ p_{mk} = \frac{1}{N} \sum_{k=1}^{N} I(y = k) $ </center>

## <center> $ H(Q_{m}) = -\sum_{k=1}^{N} p_{mk} \cdot \log(p_{mk}) $ </center> 

```python
entropy = np.copy(P) * 0
s = 0

for i, p in enumerate(P):
    s = p * np.log(p)
    entropy[i] = -s
```

```python
ax = sns.lineplot(
    x = P,
    y = entropy
)

ax.set(
    title = "Log Loss Entropy",
    xlabel = "Probability",
    ylabel = "Liklihood"
)

plt.show()
```

## <center> Gini Impurity </center> 

## <center> $ p_{mk} = \frac{1}{N} \sum_{k=1}^{N} I(y = k) $ </center>

## <center> $ H(Q_{m}) = \sum_{k=1}^{N} p_{mk} \cdot (1 - p_{mk}) $ </center>

```python
gini = np.copy(P) * 0
s = 0

for i, p in enumerate(P):
    s = p * (1 - p)
    gini[i] = s
```

```python
ax = sns.lineplot(
    x = P,
    y = gini
)

ax.set(
    title = "Gini",
    xlabel = "Probability",
    ylabel = "Liklihood"
)

plt.show()
```
