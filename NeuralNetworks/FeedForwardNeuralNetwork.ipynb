## <center> Approximate a Function </center> 

## <center> $ f(x) = sin(2 \cdot x) $ </center>

```python
N = 1000
x = np.linspace(-6, 6, N)

y_true = np.sin(2 * x)
a = y_true + np.random.normal(0, 0.1, N)
```

```python
plt.plot(x, y_true, color = "black", linewidth = 2, label = "True Function")
plt.scatter(x, a, color = "red", marker = ".", label = "Observed Points")

plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.legend(loc = "best")
plt.show()
```

```python
r = 0.1
w = np.random.uniform(0, 1, (N, N))
b = 0

z = np.dot(w, a)
y_pred = z
delta = ()

epochs = 10
err = []
r = 0.1

for i in range(epochs):
    
    err.append(np.sum((y_true - y_pred)**2) / N)
```
