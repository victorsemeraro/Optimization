# <center> <u> Feed Forward Neural Network </u> </center> 

```python
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(
    context = "notebook",
    style = "darkgrid", 
    palette = "bright",
)

sns.set(rc = {'figure.figsize':(12, 6)})
```

## <center> Approximate a Function </center> 

## <center> $ f(x) = sin(2 \cdot x) $ </center>

```python
N = 100
x = np.linspace(-6, 6, N)

y_true = np.array([np.sin(2 * x)])
a = y_true 
```

```python
plt.plot(x, y_true[0], color = "black", linewidth = 2, label = "True Function")
plt.scatter(x, a[0], color = "red", marker = ".", label = "Observed Points")

plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.legend(loc = "best")
plt.show()
```

## <center> Activation Function </center> 

## <u> Linear </u> 

## <center> $ f(x) = x $ </center>

## <center> $ f'(x) = 1 $ </center>

```python
linear = x
linear_d = np.ones(100)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 6))

ax1.plot(x, linear)
ax1.set_title("Linear")

ax2.plot(x, linear_d)
ax2.set_title("Linear Derivative")

plt.show()
```

## <center> Mathematical Structure </center>

## <center> $ W = (n, a) $ </center>

## <center> $ A = (a, m) $ </center>

## <center> $ b = (n, 1) $ </center>

```python
r = 0.1
W = np.random.randn(1, 1)
b = np.zeros((1, 1))
```

```python
print("A: ", a.shape)

print("W: ", W.shape)

print("b: ", b.shape)
```

## <center> Propagation </center>

## <center> $ z = WA + b $ </center>

## <center> $ z = (n, a) (a, m) + (n, 1) $ </center> 

## <center> $ z = (n, m) $ </center>

## <center> $ A = g(z) $ </center>

```python
z = np.dot(W, a) + b

print("Z: ", z.shape)

y_pred = z
```

```python
plt.plot(x, y_pred[0], label = "First Propagation")
plt.plot(x, y_true[0], label = "True Value")

plt.legend(loc = "best")
plt.show()
```

## <center> Objective Function </center>

## <center> $ C = \frac{1}{N} \sum_{i=1}^{N} L(y_{i}, \hat{y_{i}}) $ </center>

## <center> $ C = \frac{1}{N} \sum_{i=1}^{N} (g(z) - \hat{y_{i}})^{2} $ </center>

## <center> $ C = \frac{1}{N} \sum_{i=1}^{N} (g(WA + b) - \hat{y_{i}})^{2} $ </center>

## <center> Backpropagation </center> 

## <center> $ dz = g'(z) \cdot \nabla(a) = (g(WA + b) - \hat{y_{i}}) \cdot g'(z)$ </center> 

## <center> $ dw = \frac{1}{N} \cdot (dz \cdot a_{i-1}) $ </center>

## <center> $ db = \frac{1}{N} \sum_{i=1}^{N} (dz) $ </center>

## <center> $ da = W \cdot dz $ </center>

```python
err = 1 / 100 * (y_pred[0] - y_true[0])**2
dA = y_pred[0] - y_true[0]

dZ = np.multiply(np.ones((1, 100)), dA)
dW = 1 / 100 * np.dot(dZ, a.T)
dB = 1 / 100 * np.sum(dZ, axis = 1)

dA_prev =  np.dot(W.T, dZ)
```

## <center> Gradient Descent </center> 

## <center> $ w_{i+1} = w_{i} - \gamma dw $ </center>

## <center> $ b_{i+1} = b_{i} - \gamma db $ </center>

```python
W = W - r * dW
b = b - r * dB
```

```python
z = np.dot(W, a) + b

y_pred = z
```

```python
plt.plot(x, y_pred[0], label = "Second Propagation")

plt.plot(x, y_true[0], label = "True Value")

plt.legend(loc = "best")
plt.show()
```

```python
cost = []

for i in range(50):
    
    dZ = np.multiply(np.ones((1, 100)), dA)
    dW = 1 / 100 * np.dot(dZ, a.T)
    dB = 1 / 100 * np.sum(dZ, axis = 1)

    W = W - r * dW
    b = b - r * dB

    dA_prev =  np.dot(W.T, dZ)
    
    z = np.dot(W, a) + b
    y_pred = z
    
    cost.append(np.sum(1 / 100 * (y_pred[0] - y_true[0])**2))
    dA = y_pred[0] - y_true[0]
```

```python
plt.plot(cost)

plt.title("Cost Function")
plt.show()
```
