# <center> <u> Neural Networks Activation Functions </u> </center> 

```python
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(
    context = "notebook",
    style = "darkgrid", 
    palette = "bright",
)

sns.set(rc = {'figure.figsize':(12, 6)})
```

## <u> Propagation </u> 

## <center> $ z^{(l)} = W{(l)} \cdot a^{l-1} + b^{(l)} $ </center>

## <u> Activation Functions </u> 

## <center> $ a^{(l)} = g^{(l)} (z^{(l)}) = g^{(l)} (W{(l)} \cdot a^{l-1} + b^{(l)}) $ </center>

```python
x = np.linspace(-6, 6, 1000)
```

## <u> RELU </u>

## <center> $ f(x) = \max(0, x) $ </center>

## <center> $ f'(x) = \max(0, \frac{x}{|x|}) $ </center>

```python
relu = np.maximum(x, 0)
relu_d = np.maximum((x / np.abs(x)), 0)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 6))

ax1.plot(x, relu)
ax1.set_title("RELU")

ax2.plot(x, relu_d)
ax2.set_title("RELU Derivative")

plt.show()
```

## <u> Swish </u> 

## <center> $ f(x) = \frac{x}{1 + e^{-x}} $ </center>

## <center> $ f'(x) = \frac{1 + e^{-x} + x \cdot e^{-x}}{(1 + e^{-x})^{2}} $ </center>

```python
swish = x / (1 + np.exp(-x))
swish_d = (1 + np.exp(-x) + x * np.exp(-x)) / (1 + np.exp(-x))**2
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 6))

ax1.plot(x, swish)
ax1.set_title("SWISH")

ax2.plot(x, swish_d)
ax2.set_title("SWISH Derivative")

plt.show()
```

## <u> Sigmoid </u> 

## <center> $ f(x) = \frac{1}{1 + e^{-x}} $ </center>

## <center> $ f'(x) = \frac{e^{-x}}{(1 + e^{-x})^{2}} $ </center>

```python
sigmoid = 1 / (1 + np.exp(-x))
sigmoid_d = np.exp(-x) / (1 + np.exp(-x))**2
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 6))

ax1.plot(x, sigmoid)
ax1.set_title("Sigmoid")

ax2.plot(x, sigmoid_d)
ax2.set_title("Sigmoid Derivative")

plt.show()
```

## <u> Hyperbolic Tangent </u> 

## <center> $ f(x) = \tanh(x) $ </center>

## <center> $ f'(x) = \frac{1}{cosh^{2}(x)} $ </center>

```python
tanh = np.tanh(x)
tanh_d = 1 / np.cosh(x)**2
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 6))

ax1.plot(x, tanh)
ax1.set_title("Hyperbolic Tangent")

ax2.plot(x, tanh_d)
ax2.set_title("Sigmoid Derivative")

plt.show()
```

## <u> SoftMax </u> 

## <center> $ f(x) = \frac{e^{x}}{\sum_{i=1}^{N}e^{x}} $ </center>

## <center> $ f'(x) = $ </center> 

```python
softmax = np.exp(x) / np.sum(np.exp(x))
softmax_d = np.exp(x) / np.sum(np.exp(x))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 6))

ax1.plot(x, softmax)
ax1.set_title("Softmax")

ax2.plot(x, softmax_d)
ax2.set_title("Softmax Derivative")

plt.show()
```
