```python
import numpy as np
import pandas as pd 
from numba import jit, prange

import seaborn as sns
import matplotlib.pyplot as plt 
```

## <center> Set Seaborn Theme </center>

```python
sns.set_theme(
    style = "darkgrid",
    palette = "bright",
)

plt.rc('figure', figsize = (10, 5))
```

## <center> Radial Basis Kernel </center> 

## <center> $ K(x, x') =  e^{\frac{-||x - x'||^{2}}{2 \sigma^2}} $ </center>

```python
@jit(nopython = True)
def radial_basis_kernel(x):
    """
    
    """
    
    return np.exp(-x**2 / 2)
```

```python
N = 1000
x = np.linspace(0, 5, N)
y = radial_basis_kernel(x)
```

```python
ax = sns.lineplot(
    x = x,
    y = y,
)

ax.set(
    title = "Radial Basis Kernel",
)

plt.show()
```
