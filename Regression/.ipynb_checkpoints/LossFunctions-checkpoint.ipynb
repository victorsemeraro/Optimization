# <center> <u> Loss Functions </u> </center>

## 1. <u> Squared Error Loss </u> 

#### - Highly sensitive to outliers 

## 2. <u> Log Cosh Loss </u> 

#### - Differentiable, approximately linear 

## 3. <u> Huber Loss </u> 

#### - Outlier threshold parameter

## 4. <u> Squared Logarithmic Error Loss </u> 

#### - Concave outliers

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

## <center> Set Seaborn Theme </center>

```python
sns.set_theme(
    style = "darkgrid",
    palette = "bright",
)

plt.rc('figure', figsize = (10, 5))
```

```python
N = 1000
x = np.linspace(-10, 10, N)
```

## <center> Squared Error </center> 

## <center> $ L(\theta) = \frac{1}{N} \sum_{i=1}^{N} (x_{i} - \hat{x})^{2} $ </center>

```python
def squared_error(x):
    """
    
    """
    
    return x**2
```

```python
ax = sns.lineplot(
    x = x,
    y = squared_error(x),
)

ax.set(
    title = "Squared Error Loss",
    xlabel = "X", 
    ylabel = "Y",
)

plt.show()
```

## <center> Log Cosh Loss </center> 

## <center> $ L(\theta) = \frac{1}{N} \sum_{i=1}^{N} log(cosh(x_{i} - \hat{x})) $ </center>

```python
def log_cosh(x):
    """
    
    """
    
    return np.log(np.cosh(x))
```

```python
ax = sns.lineplot(
    x = x,
    y = log_cosh(x),
)

ax.set(
    title = "Log Cosh Loss",
    xlabel = "X", 
    ylabel = "Y",
)

plt.show()
```

## <center> Huber Loss </center> 

## <center> $ L(\theta) = \begin{cases}
        \frac{1}{N} \sum_{i=1}^{N} \frac{1}{2} (x_{i} - \hat{x})^2 & \text{if } |x_{i} - \hat{x}| \leq \delta \\
        \frac{1}{N} \sum_{i=1}^{N} \delta (|x_{i} - \hat{x}| - \frac{1}{2} \delta) & \text{if } x \gt \delta
    \end{cases} $ </center> 

```python
def huber_loss(x, delta):
    """
    
    """
    
    return [(delta * (np.abs(a) - delta / 2)) if np.abs(a) > delta else (a**2 / 2) for a in x]
```

```python
delta = 2

ax = sns.lineplot(
    x = x,
    y = huber_loss(x, delta),
)

ax.set(
    title = "Huber Loss",
    xlabel = "X", 
    ylabel = "Y",
)

plt.show()
```

## <center> Squared Logarithmic Error </center> 

## <center> $ L(\theta) = \frac{1}{N} \sum_{i=1}^{N} log((x_{i} - \hat{x})^{2} + 1) $ </center>

```python
def squared_logarithmic_error(x):
    """
    
    """
    
    return np.log(x**2 + 1)
```

```python
ax = sns.lineplot(
    x = x,
    y = squared_logarithmic_error(x),
)

ax.set(
    title = "Squared Logarithmic Error Loss",
    xlabel = "X", 
    ylabel = "Y",
)

plt.show()
```

## <center> All Together </center> 

```python
ax = sns.lineplot(
    x = x,
    y = squared_error(x),
    label = "Squared",
)

sns.lineplot(x = x, y = log_cosh(x), label = "Log Cosh")
sns.lineplot(x = x, y = huber_loss(x, delta), label = "Huber")
sns.lineplot(x = x, y = squared_logarithmic_error(x), label = "Squared Log")

ax.set(
    title = "Squared Error Loss",
    xlabel = "X", 
    ylabel = "Y",
)

plt.show()
```
